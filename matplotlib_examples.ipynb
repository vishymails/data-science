```python
import matplotlib.pyplot as plt 

%matplotlib inline 

import numpy as np


x= np.arange(0, 10)
y = np.arange(11, 21)

a= np.arange(40, 50)
b = np.arange(50, 60)
```

```python
plt.scatter(x, y, c="g")

plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.title('Graph in 2D')

plt.savefig('Test.png')
```

```python
y = x* x

plt.plot(x, y, 'r*', linestyle='dashed', linewidth=2, markersize=12)

plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.title('2D diagram')
```

```python
plt.subplot(2,2,1)
plt.plot(x, y,'r--')

plt.subplot(2,2,2)
plt.plot(x, y,'g*--')

plt.subplot(2,2,3)
plt.plot(x, y,'bo')


plt.subplot(2,2,4)
plt.plot(x, y,'go')
```

```python
x = np.arange(1,11)

y = 3 * x + 5

plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.title('2D diagram 2')

plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
plt.title("sin wave form ")

plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 5 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2, 1, 1)
plt.plot(x, y_sin, 'r--')
plt.title("sin wave form ")


plt.subplot(2, 1, 2)
plt.plot(x, y_cos, 'g--')
plt.title("cosine wave form ")

plt.show()
```

```python

x = [2,8, 10]
y = [11, 16, 9]

x2 = [3,9,11]
y2 = [6, 15, 7]

plt.bar(x, y)
plt.bar(x2, y2, color = 'g')

plt.title("Bar Graph")

plt.xlabel('X axis')
plt.ylabel("Y axis ")

plt.show()
```
