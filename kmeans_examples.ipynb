```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('Customers.csv')

X = dataset.iloc[:, [3,4]].values 

from sklearn.cluster import KMeans 

```

```python
wcss = []

for i in range(1, 11): 
    kmeans = KMeans(n_clusters=i, init = 'k-means++', random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)


plt.plot(range(1, 11), wcss)
plt.title("The Elbow Method ")
plt.xlabel("Numbers of clusters ")
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init = 'k-means++', random_state=0)
y_kmeans= kmeans.fit_predict(X)

print(y_kmeans)
```

```python
plt.scatter(X[y_kmeans==0,0], X[y_kmeans==0,1], s=100, c='red', label = "mycluster1")
plt.scatter(X[y_kmeans==1,0], X[y_kmeans==1,1], s=100, c='blue', label = "mycluster2")
plt.scatter(X[y_kmeans==2,0], X[y_kmeans==2,1], s=100, c='green', label = "mycluster3")
plt.scatter(X[y_kmeans==3,0], X[y_kmeans==3,1], s=100, c='cyan', label = "mycluster4")
plt.scatter(X[y_kmeans==4,0], X[y_kmeans==4,1], s=100, c='magenta', label = "mycluster5")
plt.show()
```

```python
plt.scatter(X[y_kmeans==0,0], X[y_kmeans==0,1], s=100, c='red', label = "mycluster1")
plt.scatter(X[y_kmeans==1,0], X[y_kmeans==1,1], s=100, c='blue', label = "mycluster2")
plt.scatter(X[y_kmeans==2,0], X[y_kmeans==2,1], s=100, c='green', label = "mycluster3")
plt.scatter(X[y_kmeans==3,0], X[y_kmeans==3,1], s=100, c='cyan', label = "mycluster4")
plt.scatter(X[y_kmeans==4,0], X[y_kmeans==4,1], s=100, c='magenta', label = "mycluster5")




plt.scatter (kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = "yellow", label = 'Centroids')

plt.title("Clusters of customers ")
plt.xlabel("Annual Income in Rs ")
plt.ylabel('Spending Score (1 -100)')
plt.legend()
plt.show()
```
